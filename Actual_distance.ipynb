```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "https://raw.githubusercontent.com/Bismaimtiazz/Planet-data/main/Venus.csv"
venus_data = np.genfromtxt(path, skip_header=60, skip_footer = 80, delimiter=',', names=['Date__(UT)__HR:MN','a','b', 'delta', 'deldot'])
```

```python
path =  'https://raw.githubusercontent.com/Bismaimtiazz/Planet-data/main/Earth.csv'
earth_data = np.genfromtxt(path, skip_header=67, skip_footer = 80, delimiter=',', names=['Date__(UT)__HR:MN','a','b', 'delta', 'deldot'])
```

```python
path = 'https://raw.githubusercontent.com/Bismaimtiazz/Planet-data/main/Mars.csv'
mars_data = np.genfromtxt(path, skip_header=62, skip_footer = 80, delimiter=',', names=['Date__(UT)__HR:MN','a','b', 'delta', 'deldot'])
```

```python
path = 'https://raw.githubusercontent.com/Bismaimtiazz/Planet-data/main/Ceres.csv'
ceres_data = np.genfromtxt(path, skip_header=74, skip_footer = 80, delimiter=',', names=['Date__(UT)__HR:MN','a','b', 'delta', 'deldot'])
```

```python
path = 'https://raw.githubusercontent.com/Bismaimtiazz/Planet-data/main/Jupiter.csv'
jupiter_data = np.genfromtxt(path, skip_header=61, skip_footer = 80, delimiter=',', names=['Date__(UT)__HR:MN','a','b', 'delta', 'deldot'])
```

```python
path = 'https://raw.githubusercontent.com/Bismaimtiazz/Planet-data/main/Saturn.csv'
saturn_data = np.genfromtxt(path, skip_header=61, skip_footer = 80, delimiter=',', names=['Date__(UT)__HR:MN','a','b', 'delta', 'deldot'])
```

```python
path = 'https://raw.githubusercontent.com/Bismaimtiazz/Planet-data/main/Uranus.csv'
uranus_data = np.genfromtxt(path, skip_header=61, skip_footer = 80, delimiter=',', names=['Date__(UT)__HR:MN','a','b', 'delta', 'deldot'])
```

```python
path = 'https://raw.githubusercontent.com/Bismaimtiazz/Planet-data/main/Pluto.csv'
pluto_data = np.genfromtxt(path, skip_header=57, skip_footer = 80, delimiter=',', names=['Date__(UT)__HR:MN','a','b', 'delta', 'deldot'])
```

```python
distances = [venus_data['delta'], earth_data['delta'], mars_data['delta'], ceres_data['delta'], jupiter_data['delta'], saturn_data['delta'], uranus_data['delta'], pluto_data['delta']]
```

```python
# Data for the distances of each planet from the Sun (in AU)
distances = np.array([0.72, 1.00, 1.52, 2.77, 5.20, 9.58, 19.18, 39.53])
```

```python
# Smooth curve with discrete points
x = np.linspace(0, 7, 1000)
y = np.interp(x, np.arange(8), distances)
```

```python
# Plot the data
plt.plot(x, y)
plt.plot(np.arange(8), distances, 'o')
for i, distance in enumerate(distances):
    plt.text(i, distance+0.2, f'({i},{distance})', ha='center')

plt.xticks(np.arange(8), ['Venus', 'Earth', 'Mars', 'Ceres', 'Jupiter', 'Saturn', 'Uranus', 'Pluto'])
plt.xlabel('Celestial Body')
plt.ylabel('Distance from Sun (AU)')
plt.title('Actual Distance of Planets from the Sun')
plt.grid()

plt.show()
```
